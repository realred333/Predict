```python
# 데이터 전처리(MinMaxScaler을 이용해서 모든 값을 0에서 1사이의 값으로 바꾼다.)

from sklearn.preprocessing import MinMaxScaler

#전처리 스케일링
scaler=MinMaxScaler()
scale_cols=['Open','High','Low','Close','Volume']
scaled=scaler.fit_transform(df_chart[scale_cols])
df_scaled=pd.DataFrame(scaled,columns=scale_cols)
```

```python
df_scaled
```

```python
print(f'data 개수 : {len(df_scaled["Close"])}')

#앞의 2000개 일봉을 train, 나머지는 valid
split=2000
train = df_scaled['Close'][0:2000]
valid = df_scaled['Close'][2000:]
```

```python
# windows_size의 개수만큼 확인해서 다음 값이 나오는지 확인 
# 그리고 한칸씩 옆으로 가면서 반복하는 작업 준비

import numpy as np
def window_data(data, window_size):
    feature = []
    label = []
    for i in range(len(data)-window_size):
        feature.append(np.array(data.iloc[i:i+window_size]))
        label.append(np.array(data.iloc[i+window_size]))
    return np.array(feature), np.array(label)

window_size=20
train_x, train_y = window_data(train, window_size)
print(f'TRAIN_X : (batch_size, window_size): {train_x.shape}')
print(f'TRAIN_Y : (batch_size, window_size): {train_y.shape}')
test_x, test_y = window_data(valid, window_size)
```

```python
#모델 빌드!
import tensorflow as tf
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Lambda(lambda x : tf.expand_dims(x,axis=-1)),
    tf.keras.layers.Conv1D(filters=32, kernel_size=8,
           padding="causal",
           activation="relu",
           input_shape=[window_size,1]),
    tf.keras.layers.LSTM(32, return_sequences=False),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(1)])
```

```python
#learn schedule를 이용해서 최적의 학습값을 찾는다

lr_schedule = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-4 * 10**(epoch / 20))
optimizer = tf.keras.optimizers.SGD(lr=1e-4, momentum=0.9)

model1.compile(loss=tf.keras.losses.Huber(), optimizer=optimizer, metrics=['mse'])

history = model1.fit(train_x,train_y,
                     validation_data=(test_x,test_y),
                     epochs=100,
                     callbacks=[lr_schedule])

#적정 learning rate 확인
plt.semilogx(history.history["lr"], history.history["loss"])
plt.axis([1e-4, 1, 0, 0.003])
plt.title('Loss @ learning rate')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, Lambda
from tensorflow.keras.losses import Huber
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint


model = Sequential([
    # 1차원 feature map 생성
    tf.keras.layers.Lambda(lambda x : tf.expand_dims(x,axis=-1)),
    Conv1D(filters=32, kernel_size=8,
           padding="causal",
           activation="relu",
           input_shape=[window_size, 1]),
    # LSTM
    LSTM(32, activation='tanh'),
    Dense(16, activation="relu"),
    Dense(1),
])

# Sequence 학습에 비교적 좋은 퍼포먼스를 내는 Huber()를 사용합니다.
loss = Huber()
optimizer = Adam(0.0005)
model.compile(loss=Huber(), optimizer=optimizer, metrics=['mse'])

# earlystopping은 10번 epoch통안 val_loss 개선이 없다면 학습을 멈춥니다.
earlystopping = EarlyStopping(monitor='val_loss', patience=10)

# val_loss 기준 체크포인터도 생성합니다.
filename = os.path.join('tmp', 'ckeckpointer.ckpt')
checkpoint = ModelCheckpoint(filename, 
                             save_weights_only=True, 
                             save_best_only=True, 
                             monitor='val_loss', 
                             verbose=1)
history = model1.fit(train_x,train_y,
                     validation_data=(test_x,test_y),
                     epochs=500,
                     callbacks=[checkpoint, earlystopping])
```

```python
#pred 예측값 생성
pred = model1.predict(test_x)

#시각화
plt.figure(figsize=(12, 9))
plt.plot(np.asarray(test_y), label='actual')
plt.plot(pred, label='prediction')
plt.title('BTC USDT Prediction')
plt.legend()
plt.show()
```
